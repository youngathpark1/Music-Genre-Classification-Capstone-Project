# Convolutional Recurrent Neural Network

Based on some publications that I came across during my research, I found that combining convolutional layers with recurrent layers is gaining attraction in the field of machine learning and sound analysis such as speech, voice, and music recongition. This method is certainly appropriate to apply for this project since Convolutional Neural Networks perform really well with image and spatial data and Recurrent Neural Networks have shown to work really well with sequential data. Mel-Spectrograms are images with temporal information, which is sequential in time, so I decided to build out a CRNN model and measure its performance based on accuracy.

```python
# Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, GRU, TimeDistributed, LSTM, Permute, Reshape
from tensorflow.keras.utils import to_categorical, plot_model
```

```python
# Attach google drive

from google.colab import drive 
drive.mount('/content/gdrive')
```

```python
# Function to save numpy array and rest of other columns seperately and then brings them together for later use

def create_df(np_path, df_path):

  loaded_np = np.load(np_path, allow_pickle=True) # Load numpy arrays
  df_exported = pd.read_csv(df_path) # Read in rest of dataframe
  df_exported['melspecs'] = loaded_np # Merge

  return df_exported
```

```python
np_path = '/content/gdrive/MyDrive/Music-Genre-Project/melspecs/data.npy'
df_path = '/content/gdrive/MyDrive/Music-Genre-Project/melspecs/df.csv'
```

```python
df = create_df(np_path, df_path)
df.head()
```

# Convolutional Recurrent Neural Network

```python
# Reshaping mel-spectrograms to be suitable for CNN

X = np.concatenate(np.array(df['melspecs'])).reshape(df['melspecs'].shape[0], 128, 130, 1)
y = df['genre_num']

# Standardize X
X /= 255

# one hot encode y
y = to_categorical(y)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
# Construct the architecture

# Instantiate
crnn_model = Sequential()

# Conv layer
crnn_model.add(Conv2D(filters=16,
                     padding='same',
                     kernel_size=(3,3),
                     activation='relu',
                     input_shape=(128, 130, 1)))

# Max pooling
crnn_model.add(MaxPooling2D(pool_size=(2,2)))

# Conv layer
crnn_model.add(Conv2D(filters=32,
                      padding='same',
                     kernel_size=(3,3),
                     activation='relu'))

# Max pooling
crnn_model.add(MaxPooling2D(pool_size=(2,2)))

# Conv layer
crnn_model.add(Conv2D(filters=64,
                      padding='same',
                     kernel_size=(3,3),
                     activation='relu'))

# Max pooling
crnn_model.add(MaxPooling2D(pool_size=(2,2)))

# Conv layer
crnn_model.add(Conv2D(filters=128,
                      padding='same',
                     kernel_size=(3,3),
                     activation='relu'))

# Max pooling
crnn_model.add(MaxPooling2D(pool_size=(2,2)))

# Conv layer
crnn_model.add(Conv2D(filters=64,
                      padding='same',
                     kernel_size=(3,3),
                     activation='relu'))

# Max pooling
crnn_model.add(MaxPooling2D(pool_size=(2,2)))

# Flatten
crnn_model.add(Flatten())

# Reshape
crnn_model.add(Reshape((-1, 128)))

# GRU
crnn_model.add(GRU(128, return_sequences=True))
crnn_model.add(GRU(64, return_sequences=True))
crnn_model.add(GRU(32, return_sequences=False))


# Dense
crnn_model.add(Dense(128, activation='relu'))
crnn_model.add(Dense(64, activation='relu'))
crnn_model.add(Dense(32, activation='relu'))

# Outer layer
crnn_model.add(Dense(10, activation='softmax'))

# Compile
crnn_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])

# Fit
crnn_history = crnn_model.fit(X_train, y_train, epochs=150, batch_size=512, validation_data=(X_test, y_test))
```

```python
# Graph Loss and Accuracy

train_loss = crnn_history.history['loss']
test_loss = crnn_history.history['val_loss']

train_acc = crnn_history.history['acc']
test_acc = crnn_history.history['val_acc']


plt.figure(figsize=(20,8))
plt.subplot(1, 2, 1)
plt.xlabel(xlabel='Epochs', fontsize=12)
plt.ylabel(ylabel='Loss', fontsize=12)
plt.title('CNN Model Loss', fontsize=15)
plt.plot(train_loss, label='Training Loss')
plt.plot(test_loss, label='Testing Loss')
plt.legend(fontsize=15)

plt.subplot(1, 2, 2)
plt.xlabel(xlabel='Epochs', fontsize=12)
plt.ylabel(ylabel='Accuracy', fontsize=12)
plt.title('CNN Model Accuracy', fontsize=15)
plt.plot(train_acc, label='Training Accuracy')
plt.plot(test_acc, label='Testing Accuracy')
plt.legend(fontsize=15);
```

Without any regularization, CNN model is definitely overfit. You can start to see the divergence between training and testing loss and accuracy starting around 20 epochs.

```python
plot_model(crnn_model, show_layer_names=True, show_shapes=True)
```

# Convolutional Recurrent Neural Network with Dropouts

```python
# Construct the architecture

# Instantiate
crnn_model_dropout = Sequential()

# Conv layer
crnn_model_dropout.add(Conv2D(filters=16,
                     kernel_size=(3,3),
                     padding='same',
                     activation='relu',
                     input_shape=(128, 130, 1)))

# Max pooling
crnn_model_dropout.add(MaxPooling2D(pool_size=(2,2)))

# Conv layer
crnn_model_dropout.add(Conv2D(filters=32,
                     padding='same',
                     kernel_size=(3,3),
                     activation='relu'))

# Max pooling
crnn_model_dropout.add(MaxPooling2D(pool_size=(2,2)))

# Conv layer
crnn_model_dropout.add(Conv2D(filters=64,
                     padding='same',
                     kernel_size=(3,3),
                     activation='relu'))

# Max pooling
crnn_model_dropout.add(MaxPooling2D(pool_size=(2,2)))

# Conv layer
crnn_model_dropout.add(Conv2D(filters=128,
                     padding='same',
                     kernel_size=(3,3),
                     activation='relu'))

# Max pooling
crnn_model_dropout.add(MaxPooling2D(pool_size=(2,2)))

# Conv layer
crnn_model_dropout.add(Conv2D(filters=64,
                     padding='same',
                     kernel_size=(3,3),
                     activation='relu'))

# Max pooling
crnn_model_dropout.add(MaxPooling2D(pool_size=(2,2)))

# Flatten
crnn_model_dropout.add(Flatten())

# Reshape
crnn_model_dropout.add(Reshape((-1, 128)))

# GRU
crnn_model_dropout.add(GRU(128, return_sequences=True))
crnn_model_dropout.add(GRU(64, return_sequences=False))


# Dense
crnn_model_dropout.add(Dense(128, activation='relu'))
crnn_model_dropout.add(Dropout(0.2))
crnn_model_dropout.add(Dense(64, activation='relu'))
crnn_model_dropout.add(Dropout(0.2))
crnn_model_dropout.add(Dense(32, activation='relu'))
crnn_model_dropout.add(Dropout(0.2))

# Outer layer
crnn_model_dropout.add(Dense(10, activation='softmax'))

# Compile
crnn_model_dropout.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])

# Fit
crnn_dropout_history = crnn_model_dropout.fit(X_train, y_train, epochs=150, batch_size=512, validation_data=(X_test, y_test))
```

```python
# Graph Loss and Accuracy

train_loss_dropout = crnn_dropout_history.history['loss']
test_loss_dropout = crnn_dropout_history.history['val_loss']

train_acc_dropout = crnn_dropout_history.history['acc']
test_acc_dropout = crnn_dropout_history.history['val_acc']


plt.figure(figsize=(20,8))
plt.subplot(1, 2, 1)
plt.xlabel(xlabel='Epochs', fontsize=12)
plt.ylabel(ylabel='Loss', fontsize=12)
plt.title('CNN Model Loss', fontsize=15)
plt.plot(train_loss_dropout, label='Training Loss')
plt.plot(test_loss_dropout, label='Testing Loss')
plt.legend(fontsize=15)

plt.subplot(1, 2, 2)
plt.xlabel(xlabel='Epochs', fontsize=12)
plt.ylabel(ylabel='Accuracy', fontsize=12)
plt.title('CNN Model Accuracy', fontsize=15)
plt.plot(train_acc_dropout, label='Training Accuracy')
plt.plot(test_acc_dropout, label='Testing Accuracy')
plt.legend(fontsize=15);
```

```python
plot_model(crnn_model_dropout, show_layer_names=True, show_shapes=True)
```
